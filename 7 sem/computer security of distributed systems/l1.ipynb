Vigenere Cipher Realisation:

```python
def vigenere_cipher(plaintext, key):
    key_length = len(key)
    key_as_int = [ord(i) - 97 for i in key]
    plaintext_int = [ord(i) - 97 for i in plaintext]
    ciphertext = ''
    counter = 0
    for i in range(len(plaintext_int)):
        if(plaintext_int[i] >= -32 and plaintext_int[i] <= -7):
            plaintext_int[i] += 32
        if(plaintext_int[i] >= 0 and plaintext_int[i] <= 26):
            value = (plaintext_int[i] + key_as_int[(i - counter) % key_length]) % 26
            ciphertext += chr(value + 97)
        else:
            counter+=1
    return ciphertext
```

```python
def find_deltas(text, l_gramm_length):
    dictionary = {}
    entries = {}
    deltas = []
    for i in range((len(text) - l_gramm_length + 1)):
        curr_l_gramm = text[i:i + l_gramm_length]
        if dictionary.get(curr_l_gramm) is None:
            dictionary[curr_l_gramm] = 1
            entries[curr_l_gramm] = i
        else:
            dictionary[curr_l_gramm] += 1
            deltas.append(i - entries[curr_l_gramm])
            entries[curr_l_gramm] = i
    return deltas
```

```python
def array_gcd(array):
    val = 1
    for j in range(2,51):
        counter = 0
        for i in range(len(array)):
            if(array[i] % j == 0):
                counter+=1
        if(counter/len(array) >= 0.9 ):
            val = j
    return val
```

```python
def kasiski(str):
    deltas_arr = []
    for i in range(3, len(str)):
        deltas = find_deltas(str, i)
        for j in range(len(deltas)):
            deltas_arr.append(deltas[j])
    if(len(deltas_arr) >=2):
        l = array_gcd(deltas_arr)
    else:
        return 1
    return l
```
